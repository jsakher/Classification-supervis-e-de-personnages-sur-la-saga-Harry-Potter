# Loyalty

Extraction of all the loyalties:

```python
df_loyalty = data_char[['Character', 'Loyalty']].copy()
df_loyalty.Loyalty = df_loyalty.Loyalty.fillna('')
df_loyalty.Loyalty = df_loyalty.Loyalty.apply(lambda t : t.lower().split(', '))
```

```python
loyalty_col = []
for i in range(len(df_loyalty)):
    loyalty_list = df_loyalty.Loyalty[i]
    for j in range(len(loyalty_list)):
        loyalty = loyalty_list[j]
        if loyalty not in loyalty_col:
            loyalty_col.append(loyalty)
loyalty_col.remove('')
```

```python
df_loyalty_3 = data_char_3[['Character', 'Loyalty']].copy()
df_loyalty_3.Loyalty = df_loyalty_3.Loyalty.fillna('')
df_loyalty_3.Loyalty = df_loyalty_3.Loyalty.apply(lambda t : t.lower().split(', '))
```

```python
loyalty_col_3 = []
for i in range(len(df_loyalty_3)):
    loyalty_list_3 = df_loyalty_3.Loyalty[i]
    for j in range(len(loyalty_list_3)):
        loyalty_3 = loyalty_list_3[j]
        if loyalty_3 not in loyalty_col_3:
            loyalty_col_3.append(loyalty_3)
loyalty_col_3.remove('')
```

```python
loyalties = list(set(loyalty_col + loyalty_col_3))
loyalties
```

## HP1 and HP2

```python
loyalty_mat = pd.DataFrame(0, columns=loyalties, index=df_loyalty.Character).rename_axis('Character').reset_index()

for i in range(len(df_loyalty)):
    loyalty_list = df_loyalty.Loyalty[i]
    for j  in range(len(loyalties)):
        loyalty = loyalties[j]
        if loyalty in loyalty_list:
            loyalty_mat[f'{loyalty}'][i] = 1
```

## HP3

```python
loyalty_mat_3 = pd.DataFrame(0, columns=loyalties, index=df_loyalty_3.Character).rename_axis('Character').reset_index()

for i in range(len(df_loyalty_3)):
    loyalty_list_3 = df_loyalty_3.Loyalty[i]
    for j  in range(len(loyalties)):
        loyalty_3 = loyalties[j]
        if loyalty_3 in loyalty_list_3:
            loyalty_mat_3[f'{loyalty_3}'][i] = 1
```

# Skills

```python
skills_col = []
for i in range(len(df_skills)):
    skill_list = df_skills.Skills[i]
    for j in range(len(skill_list)):
        skill = skill_list[j]
        if skill not in skills_col:
            skills_col.append(skill)
skills_col.remove('')
```
